```python
import random

print('WElcome to ROCK, PAPER, SCISSORS game!')

ties = 0
wins = 0
loses = 0

while True:
    user_input = input('Enter your move: (r)ock (p)aper (s)cissors : ')
    
    if user_input == 'q' or user_input == 'Q':
        break
        


    if user_input == 'r':
        print('ROCK' + ' versus... ')
    elif user_input == 'p':
        print('PAPER' + ' versus... ')
    elif user_input == 's':
        print('SCISSORS' + ' versus... ')
    else:
        print("You know NOTHING about RockPaperScissors! That is unacceptable and you lost the game! ")
        break

    num_random = random.randint(1,3)

    if num_random == 1:
        print('ROCK')
        if user_input == 'r':
            print('It is a tie!')
            ties += 1
        elif user_input == 'p':
            print('You win!')
            wins +=1
            break
        elif user_input == 's':
            print('You lose!')
            loses += 1

    if num_random == 2:
        print('PAPER')
        if user_input == 'r':
            print('You lose!')
            loses += 1
        elif user_input == 'p':
            print('It is a tie!')
            ties += 1
        elif user_input == 's':
            print('You win!')
            wins += 1
            break



    if num_random == 3:
        print('Scissors')
        if user_input == 'r':
            print('You win!')
            wins += 1
            break
        elif user_input == 'p':
            print('You lose!')
            loses += 1
        elif user_input == 's':
            print('It is a tie!')   
            ties += 1
          
print()
print('You have ' + str(ties) + ' ties and ' + str(loses) + ' loses before your first Win.')
```
